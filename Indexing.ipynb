```python
porter_stemmer = PorterStemmer()

sentence_tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
treebank_tokenizer = TreebankWordTokenizer()

doc_dict = {}

for i in range(len(df_per_patent)):
    doc_dict[df_per_patent.iloc[i,0]] = df_per_patent.iloc[i,1]

doc_ids = list(doc_dict.keys())
print(doc_ids)
```

```python
#Read in file, tokenize into sentences, and tokenize into words
exclude = list(set(string.punctuation))
stop = list(set(stopwords.words('english')))

def query_tokenize(query):
    punkt_query = sentence_tokenizer.tokenize(query)
    filtered = [i for i in punkt_query if not pattern.search(i)]
    no_punc = []
    for i in filtered:
        punc_free = ''.join(a for a in i if a not in exclude) 
        no_punc.append(punc_free)
    sentence_words = [treebank_tokenizer.tokenize(sentence) for sentence in no_punc]
    no_stops = []
    


def read_file_and_tokenize(docs):
    tokenized_docs = []
    pattern = re.compile(r'[0-9]+.')
    for doc in docs:
        punkt_sentences = sentence_tokenizer.tokenize(doc)
        filtered = [i for i in punkt_sentences if not pattern.search(i)]
        no_punc = []
        for i in filtered:
            punc_free = ''.join(a for a in i if a not in exclude) 
            no_punc.append(punc_free)
        sentence_words = [treebank_tokenizer.tokenize(sentence) for sentence in no_punc]
        no_stops = []
        for a in sentence_words:
            no_stops_sub = []
            for b in a:
                if b not in stop:
                    stem = porter_stemmer.stem(b)
                    no_stops_sub.append(stem)
            no_stops.append(no_stops_sub)
            
        tokenized_docs.append(no_stops)
        
    tokenized_corpus = {}
    for i, doc in enumerate(tokenized_docs):
        flat_doc = [x for sentence in doc for x in sentence]
        tokenized_corpus[doc_ids[i]] = flat_doc
        #tokenized_corpus.append(flat_doc)
        
    return tokenized_corpus

token_docs = read_file_and_tokenize(doc_dict.values())
```

```python
def term_freq(corpus):
    all_terms = []
    for doc in list(corpus.values()):
        term_list = list(set(doc))
        all_terms.append(term_list)      
    
    all_terms = [x for sentence in all_terms for x in sentence]
    unique_terms = list(set(all_terms))

    tf = {}
    for i, doc in enumerate(list(corpus.values())): 
        term_freq = {}
        for word in unique_terms:
            term_freq[word] = doc.count(word)
        tf[doc_ids[i]] = term_freq
        
    return tf, unique_terms


tf, terms = term_freq(token_docs)
```

```python
tf_dict = {}
for i, doc in enumerate(list(tfdf.index)):
    vec = list(tfdf.iloc[i,:])

    dot_product = np.dot(query_vec, vec)
    norm_a = np.linalg.norm(query_vec)
    norm_b = np.linalg.norm(vec)
    cosine_similarity = dot_product / (norm_a * norm_b)
    if cosine_similarity > 0:
        tf_dict[doc] = cosine_similarity


sign_docs = list(zip(list(tf_dict.keys()),list(tf_dict.values())))
sig_df = pd.DataFrame(sign_docs, columns = ['Patent_ID', 'Cosine_Sim'])
```
